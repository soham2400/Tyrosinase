## Importing Libraries

```python
from chembl_webresource_client.new_client import new_client
```

```python
import pandas as pd
```

## Gather Data from Chembl

```python
target = new_client.target
tyrosinase = target.search("tyrosinase")
df1 = pd.DataFrame.from_dict(tyrosinase)
df1
```

```python
my_tar = df1.target_chembl_id[4]
my_tar
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id = my_tar).filter(type = "IC50")
```

```python
df_act = pd.DataFrame.from_dict(res)
df_act
```

```python
df2 = df_act[df_act.standard_value.notna()]
```

Cleaning the retrieved data by removing NA values.

## Collecting required data and pre-processing of bio-activity data

```python
selection = ["molecule_chembl_id" , "standard_value", "canonical_smiles"]
df3 = df2[selection]
```

```python
df3
```

*Compounds in the data are classified as active, inactive or intermediate based on their standard_value which is an IC50 unit. Compounds having values of less than 1000 nM will be considered to be active while those greater than 10,000 nM will be considered to be inactive. As for those values in between 1,000 and 10,000 nM will be referred to as intermediate.*

```python
bioactivity_class = []
for i in df2.standard_value:
  if float(i) <= 1000 :
    bioactivity_class.append("active")
  elif float(i) >= 10000:
    bioactivity_class.append("inactive")
  else:
    bioactivity_class.append("intermediate")
```

```python
se = pd.Series(bioactivity_class)
```

```python
ser = list(se)
```

```python
df3["bio_class"] = ser
```

```python
df3
```

## Exctracting data out to a CSV file

```python
df3.to_csv("tyrosinase_1.csv")
```
